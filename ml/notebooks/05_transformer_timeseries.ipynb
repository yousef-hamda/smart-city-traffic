# 05 · Transformer time-series (Phase-2 experiment)

A small `nn.TransformerEncoder` regressor as an alternative to the LSTM.

```python
import sys, pathlib
# Make the ml-prediction package importable from the repo root.
REPO = pathlib.Path.cwd()
while REPO != REPO.parent and not (REPO / 'ml' / 'data' / 'training.csv').exists():
    REPO = REPO.parent
sys.path.insert(0, str(REPO / 'apps' / 'ml-prediction' / 'src'))
DATA = REPO / 'ml' / 'data' / 'training.csv'
```

```python
import pandas as pd
from ml_prediction.models.congestion import CongestionModel
df = pd.read_csv(DATA)
model = CongestionModel(model_type='transformer')
model.train(df)
model.predict_speed(df)
```

Compare against the LSTM (notebook 04) on the same split. Self-attention can capture longer-range dependencies at the cost of more data.
