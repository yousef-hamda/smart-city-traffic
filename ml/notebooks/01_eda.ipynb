# 01 · Exploratory Data Analysis

Understand the synthetic Jerusalem traffic dataset before modelling. Regenerate it with `python ml/data/make_dataset.py`.

```python
import sys, pathlib
# Make the ml-prediction package importable from the repo root.
REPO = pathlib.Path.cwd()
while REPO != REPO.parent and not (REPO / 'ml' / 'data' / 'training.csv').exists():
    REPO = REPO.parent
sys.path.insert(0, str(REPO / 'apps' / 'ml-prediction' / 'src'))
DATA = REPO / 'ml' / 'data' / 'training.csv'
```

```python
import pandas as pd
df = pd.read_csv(DATA, parse_dates=[c for c in ['time','ts'] if c in pd.read_csv(DATA, nrows=0).columns])
print(df.shape)
df.head()
```

## Distributions

```python
df.describe()
```

```python
# Speed vs occupancy — the fundamental diagram should show speed
# falling as occupancy rises.
import matplotlib.pyplot as plt
if {'occupancy_pct','avg_speed_kmh'} <= set(df.columns):
    plt.scatter(df['occupancy_pct'], df['avg_speed_kmh'], s=4, alpha=0.3)
    plt.xlabel('occupancy %'); plt.ylabel('avg speed km/h'); plt.title('Fundamental diagram')
```

## Temporal pattern
Weekday rush peaks vs quiet nights/weekends.

```python
if 'hour' in df.columns:
    df.groupby('hour')['vehicle_count'].mean().plot(marker='o', title='Mean vehicle count by hour')
```
