# 06 · Evaluation & SHAP explainability

Why a prediction was made — the contributions surfaced in the dashboard's drill-down.

```python
import sys, pathlib
# Make the ml-prediction package importable from the repo root.
REPO = pathlib.Path.cwd()
while REPO != REPO.parent and not (REPO / 'ml' / 'data' / 'training.csv').exists():
    REPO = REPO.parent
sys.path.insert(0, str(REPO / 'apps' / 'ml-prediction' / 'src'))
DATA = REPO / 'ml' / 'data' / 'training.csv'
```

```python
import pandas as pd
from ml_prediction.models.demand import DemandModel, FEATURES
from ml_prediction.explain import get_top5_shap
df = pd.read_csv(DATA)
model = DemandModel(); model.train(df)
cols = [c for c in FEATURES if c in df.columns]
X = df[cols].fillna(0.0)
get_top5_shap(model._model, X, model_type='xgboost', feature_names=cols)
```

Each prediction the service returns includes these top-5 SHAP contributions so the forecast is auditable, not a black box.
