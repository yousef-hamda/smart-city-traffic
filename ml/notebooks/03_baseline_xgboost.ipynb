# 03 · Baseline — XGBoost demand forecaster

Gradient-boosted trees predicting next-hour vehicle count. The strong tabular baseline.

```python
import sys, pathlib
# Make the ml-prediction package importable from the repo root.
REPO = pathlib.Path.cwd()
while REPO != REPO.parent and not (REPO / 'ml' / 'data' / 'training.csv').exists():
    REPO = REPO.parent
sys.path.insert(0, str(REPO / 'apps' / 'ml-prediction' / 'src'))
DATA = REPO / 'ml' / 'data' / 'training.csv'
```

```python
import pandas as pd
from ml_prediction.models.demand import DemandModel
df = pd.read_csv(DATA)
model = DemandModel()
model.train(df)
preds = model.predict(df)
print('model version:', model.model_version)
preds[:10]
```

Evaluate with MAE/RMSE on a held-out split, and log to MLflow in `python -m ml_prediction.train --model xgboost`.

```python
import numpy as np
if 'vehicle_count' in df.columns:
    err = preds - df['vehicle_count'].to_numpy()
    print('MAE', float(np.mean(np.abs(err))), 'RMSE', float(np.sqrt(np.mean(err**2))))
```
