# 04 · LSTM congestion forecaster (PyTorch Lightning)

Sequence model predicting near-future average speed from a window of recent readings.

```python
import sys, pathlib
# Make the ml-prediction package importable from the repo root.
REPO = pathlib.Path.cwd()
while REPO != REPO.parent and not (REPO / 'ml' / 'data' / 'training.csv').exists():
    REPO = REPO.parent
sys.path.insert(0, str(REPO / 'apps' / 'ml-prediction' / 'src'))
DATA = REPO / 'ml' / 'data' / 'training.csv'
```

```python
import pandas as pd
from ml_prediction.models.congestion import CongestionModel
df = pd.read_csv(DATA)
model = CongestionModel(model_type='lstm')
model.train(df)
model.predict_speed(df)
```

`predict_speed` returns a mean plus an MC-dropout confidence interval — uncertainty the dashboard renders as a band.
